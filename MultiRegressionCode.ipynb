```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 500
X = np.hstack((
    np.ones((N,1)),
    np.linspace(-100, 100, N).reshape(N, 1),
    np.linspace(-1000, 1000, N).reshape(N, 1)))

```

```python
b = np.array([[25.],[5.2],[7.8]])
```

```python
Y = X@b+np.random.randn(500,1)*1000
```

```python
print('dimentions of X = {}, b = {}, Y = {}'. format(X.shape, b.shape, Y.shape))
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,1], X[:,2], Y, s=4,c='k')
```

```python
# one liner for for regression using the vectorized OLS solution / Moore Penrose PseudoInverse / Normal Equation
w= np.linalg.inv(X.T@X)@X.T@Y
```

```python
print('w = {} \n whereas \n b = {}'.format(w,b))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,1], X[:,2], Y, s=4,c='k')
ax.plot(X[:,1],X[:,2],(X@w).flatten(),'-r')
```
