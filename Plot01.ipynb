```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
X= np.linspace(-100, 100, 100).reshape(100, 1)
Y= X@np.array([[3]]) + np.array([[100]]) + np.random.randn(100,1) * 25 
```

```python
plt.scatter(X,Y)
plt.title('Regression Toy Data Set')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
W0 = (np.mean(Y)*np.mean(X**2)-(np.mean(X)*np.mean(X*Y)))/(np.mean(X**2) - (np.mean(X)**2)) 
W1 = (np.mean(X*Y) - (np.mean(X) * np.mean(Y)))/ (np.mean(X**2) - np.mean(X)**2)
```

```python
W0.shape
print(W0)
print(W1)
```

```python
Yhat = W0 + X@np.array([[W1]])
plt.scatter(X,Y)
plt.plot(X, Yhat, '-r')
```

```python
R2 = 1 - ((Y - Yhat).T@(Y - Yhat)) / (((Y - np.mean(Y)).T@(Y - np.mean(Y))))
```
