```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 1000
X = np.hstack((np.ones((N,1)), (np.random.rand(N,1)-.5)*100))
```

```python
b = np.array([[1.],[2.5],[3.7]])
```

```python
X[:]
```

```python
Y = np.hstack((X,X[:,[1]]**2))@b+np.random.randn(N,1)*1e3
```

```python
plt.figure(figsize=(7,7))
plt.scatter(X[:,[1]],Y,s=3)
plt.title('The Simulated Data with Noise')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
phi = np.hstack((X,X[:,[1]]**2))
```

```python
w = np.linalg.inv(phi.T@phi)@phi.T@Y
```

```python
Yhat = phi@w
```

```python
plt.scatter(X[:,1], Y.flatten())
plt.scatter(X[:,1],Yhat.flatten(),c='r')
```
